```python
file = '04-2010.csv'
path = 'Results/DataFrame/'
data = pd.read_csv(path+file)
most_selling_data = data[:20]
most_selling_data
```

```python
def no_whitespace(data_list):
    '''
    Example:
    data1 = no_whitespace(data['items'])
    '''
    new_data = [] 
    for data in data_list:
        new_data.append(data.replace(" ", ""))
    return new_data
```

# BACK-To-BACK bar chart

```python
plt.figure(figsize=(7,5))

color = []
for id in range(4):
    color.append('blue')
for id in range(4,len(data['items'])):
    color.append('gray')
plt.barh(most_selling_data['items'],most_selling_data['total_sale'],color = color)
# plt.set_tick_params(top='off',bottom='off',left='off',right='off', labelleft='off', labelbottom='on')
plt.title('Top selling products')
# plt.axis('off')
plt.xlabel('Currency')
plt.ylabel('Items')
plt.show()
```

```python
plt.figure()
color = []
for id,row in data.iterrows():
    if row['quantities'] > 300:
        color.append('red')
    else:
        color.append('purple')
plt.barh(most_selling_data['items'],-most_selling_data['quantities'],color = color)
plt.xlabel('Quantities')
```

# Least selling

```python
least_selling_data = data[data['quantities'] < 10] 
```

```python
title = 'Least_Selling_Items.txt'
least_selling_data_list = list(least_selling_data['items'].unique())
with open(f'{title}', 'w') as fp:
    fp.write("\n".join(str(item) for item in least_selling_data_list))
```
